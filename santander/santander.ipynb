```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
train_df.describe()
```

```python
# remove constant columns (std = 0)
remove = []
for col in train_df.columns:
    if train_df[col].std() == 0:
        remove.append(col)

train_df.drop(remove, axis=1, inplace=True)
test_df.drop(remove, axis=1, inplace=True)
```

```python
from xgboost.sklearn import XGBClassifier
```

```python
X_test = test_df.drop(['ID'], axis=1).values

X = train_df.drop(['ID', 'TARGET'], axis=1).values
y = train_df['TARGET'].values
```

```python
xgb = XGBClassifier()
```

```python
xgb.fit(X, y)
```

```python
y_pred = xgb.predict(X_test)
```

```python
pd.DataFrame(y_pred).describe()
```

```python
sub = pd.DataFrame(np.column_stack((test_df['ID'].values, y_pred)), columns=['ID', 'TARGET'])
```

```python
sub.to_csv('dummy.csv', index=False)
```
